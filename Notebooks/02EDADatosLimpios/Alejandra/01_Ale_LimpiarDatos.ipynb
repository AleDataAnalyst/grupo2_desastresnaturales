### Visualizaciones Nivel 2: variables geográficas

#### Countplots

```python
geo_cols = ['Region', 'Continent']
# Realizar countplot solo con Region y Continent,
# No Country ni Location: demasiados resultados

for col in geo_cols:
    sns.countplot(x=col, data=df)
    plt.title(f'Countplot de {col}')
    plt.xticks(rotation=75)
    plt.show()
```

#### Pie Charts

```python
geo_cols = ['Continent']

for col in geo_cols:
    category_counts = df[col].value_counts()
    labels = category_counts.index
    sizes = category_counts.values

    plt.pie(x=sizes, labels=labels, colors=palette_color, autopct='%.0f%%')
    plt.title(f'Gráfica de Pie de {col}')
    plt.legend(fontsize=6)
    plt.show()
```
