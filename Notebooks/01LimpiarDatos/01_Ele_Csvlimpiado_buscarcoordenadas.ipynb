```python
df = pd.read_csv('../../Data/03Limpio/03desastreslimpio.csv')
df.head(14)
```

```python
filter = df['Disaster Type'] == 'Earthquake'

df_earthquake = df[filter]

df_earthquake.head()
```

```python
filter = df_earthquake['Latitude'].isnull() | df_earthquake['Longitude'].isnull()

# Filtramos el dataframe para obtener solo las filas con valores nan en esas columnas
df_earthquake_nan= df_earthquake[filter]
df_earthquake_nan[['Longitude','Latitude','Location', 'Country']].head(50)
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="MozillaFirefox")
```

```python
for i, row in df_earthquake_nan.iterrows():
    if pd.isnull(row['Location']):
        location = geolocator.geocode(row['Country'])
    else:
        location = geolocator.geocode(row['Location'])
    if location:
        df_earthquake_nan.loc[i, 'Latitude'] = location.latitude
        df_earthquake_nan.loc[i, 'Longitude'] = location.longitude
        sleep(2)
```

Búsqueda con código de los nan faltantes

```python
def fill_lat_lon(row):
    if np.isnan(row['Latitude']) or np.isnan(row['Longitude']):
        location = geolocator.geocode(row['Country'])
        if location:
            row['Latitude'] = location.latitude
            row['Longitude'] = location.longitude
            sleep(1)
    return row

df_earthquake_fillnan = df_earthquake_nan.apply(fill_lat_lon, axis=1)
df_earthquake_fillnan.isnull().sum()
```

Código para limpiar coordenadas con letras

```python
def limpiar_valor(valor):
    if pd.isna(valor):
        return valor
    elif 'N' in valor or 'E' in valor:
        return valor.replace('N', '').replace('E', '')
    elif 'S' in valor or 'W' in valor:
        return '-' + valor.replace('S', '').replace('W', '')
    else:
        return valor
    
df_earthquake['Latitude'] = df_earthquake['Latitude'].apply(limpiar_valor)
df_earthquake['Longitude'] = df_earthquake['Longitude'].apply(limpiar_valor)
```
