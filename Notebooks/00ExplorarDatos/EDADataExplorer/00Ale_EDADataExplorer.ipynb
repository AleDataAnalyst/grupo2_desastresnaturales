## Bucle for para .unique y .value_counts por columna

```python
# Bucle For para obtener value counts por cada columna
columns = ['Region', 'WRI', 'Exposure', 'Vulnerability', 'Susceptibility',
           'Lack of Coping Capabilities', 'Lack of Adaptive Capacities', 'Year',
           'Exposure Category', 'WRI Category', 'Vulnerability Category',
           'Susceptibility Category']

for column in columns:
    print(f"Columna: {column}")
    print("Value Counts:")
    print(df[column].value_counts())
    print()
```

## Bucle for para .unique por columna

```python
# Bucle For para obtener value counts y unique values por cada columna
columns = ['Region', 'WRI', 'Exposure', 'Vulnerability', 'Susceptibility',
           'Lack of Coping Capabilities', 'Lack of Adaptive Capacities', 'Year',
           'Exposure Category', 'WRI Category', 'Vulnerability Category',
           'Susceptibility Category']

for column in columns:
    print(f"Columna: {column}")
    print("Unique Values:")
    print(df[column].unique())
    print()
```

## Detecci√≥n y exploraci√≥n de outliers

```python
outlier_columns = ['WRI', 'Exposure', 'Vulnerability', 'Susceptibility', 'Lack of Coping Capabilities',
                   'Lack of Adaptive Capacities']
outliers = {}
for column in outlier_columns:
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    column_outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    outliers[column] = column_outliers

# Explorar outliers
for column, outlier_data in outliers.items():
    print(f"Outliers for column: {column}")
    print(outlier_data)
    print()
```

## Rankings por columnas

```python
# Rankings por columna
rankings = {}
for column in columns:
    if df[column].dtype != 'object':
        rankings[column] = df[column].rank(ascending=False)

# Explorar rankings
for column, rank_data in rankings.items():
    print(f"Rankings for column: {column}")
    print(rank_data)
    print()
```

# Visualizaciones de outliers

## Box plots para columnas num√©ricas

```python
numerical_columns = ['WRI', 'Exposure', 'Vulnerability', 'Susceptibility', 'Lack of Coping Capabilities', 'Lack of Adaptive Capacities']
for column in numerical_columns:
    plt.figure()
    plt.boxplot(df[column])
    plt.title(f"Box Plot of {column}")
    plt.show()
```

### Barplots para columnas categ√≥ricas

```python
categorical_columns = ['Region', 'Exposure Category', 'WRI Category', 'Vulnerability Category', 'Susceptibility Category']

for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=column, data=df)
    plt.title(f"Count Plot of {column}")
    plt.xticks(rotation=45)
    plt.show()
```

### Bar plots

```python
for column in categorical_columns:
    plt.figure(figsize=(8, 8))
    df[column].value_counts().plot(kind='pie', autopct='%1.1f%%')
    plt.title(f"Pie Chart of {column}")
    plt.ylabel('')
    plt.show()
```

### Bar plots

```python
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    df[column].value_counts().plot(kind='bar', stacked=True)
    plt.title(f"Stacked Bar Plot of {column}")
    plt.xticks(rotation=45)
    plt.show()
```

### Donut plots

```python
for column in categorical_columns:
    plt.figure(figsize=(8, 8))
    df[column].value_counts().plot(kind='pie', autopct='%1.1f%%', wedgeprops=dict(width=0.3))
    plt.title(f"Donut Plot of {column}")
    plt.ylabel('')
    plt.show()
```

### Heatmaps

```python
for column in categorical_columns:
    crosstab = pd.crosstab(index=df[column], columns='count')
    plt.figure(figsize=(8, 6))
    sns.heatmap(crosstab, annot=True, cmap='Blues', cbar=False)
    plt.title(f"Heatmap de {column}")
    plt.xlabel('')
    plt.ylabel('')
    plt.show()
```
